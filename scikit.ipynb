```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

df.head()
```

```python
print(df.shape)
print(df.dtypes)
df['target'].value_counts()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.boxplot(data=df.iloc[:, :-1])
plt.title("Before Scaling")
plt.show()
```

```python
#with the box plot we can see that all 4 attributes have different ranges
#to cultivate them into having same ranges we use scaling
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.iloc[:, :-1])  # Exclude target #will scale the 4 attributes
y = df['target']
```

```python
scaled_df = pd.DataFrame(X_scaled, columns=df.columns[:-1])

print("\nAfter Scaling:\n")
print(scaled_df.describe())
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=scaled_df)
plt.title("After Scaling (StandardScaler)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2,random_state=42, stratify=y)
#x_scaled is data passed
#random_state= Each time we run the train_test_split, itll indicate the data to be split in the same way
#stratify= tells from each category we need split 20% i.e., 10 for testing, 40 for training in each category of iris
```

```python
print("Train Features Shape:", X_train.shape)
print("Train Targets Shape:", y_train.shape)
print("Test Features Shape:", X_test.shape)
print("Test Targets Shape:", y_test.shape)
```

```python
from sklearn.svm import SVC

svm_model = SVC(kernel='linear', C=1.0, random_state=42)

svm_model.fit(X_train, y_train)
#fit - this keyword is used to train any model thats there
```

```python
from sklearn.decomposition import PCA
#before there were 4 attributes for each categories of iris. As it is difficult/impossible to conduct a 4d analysis:
#we use pca. It selects the best 2 attributes and analyses them in 2d
import matplotlib.pyplot as plt
import numpy as np

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap='viridis', edgecolor='k')
plt.title("Training Data (PCA projection)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

```python
svm_pca = SVC(kernel='linear', C=1.0, random_state=42)
svm_pca.fit(X_train_pca, y_train)
```

```python
# Create a mesh for decision boundary plot
x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

Z = svm_pca.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, edgecolor='k', cmap='viridis')
plt.title("SVM Decision Boundaries (PCA projection)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

y_pred = svm_model.predict(X_test)

print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))

# Confusion Matrix
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='d',
            xticklabels=iris.target_names,
            yticklabels=iris.target_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
import streamlit as st
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier

# Load data
iris = datasets.load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
target_names = iris.target_names

st.title("Iris Flower ML Demo")

# Section: Data exploration
st.header("Data Overview")
st.dataframe(X.head())
st.write(X.describe())

# Section: Data visualization
st.header("Feature Visualization")
fig, ax = plt.subplots()
sns.boxplot(data=X, orient='h', ax=ax)
st.pyplot(fig)

# Section: Model Training and Results
st.header("Model Training")
clf = RandomForestClassifier(random_state=42)
clf.fit(X, y)
acc = clf.score(X, y)
st.write(f"Random Forest Accuracy (on all data): {acc:.2f}")

# Feature importances
st.header("Feature Importances")
importances = clf.feature_importances_
for name, imp in zip(X.columns, importances):
    st.write(f"{name}: {imp:.2f}")

fig2, ax2 = plt.subplots()
sns.barplot(x=X.columns, y=importances, ax=ax2)
plt.xticks(rotation=45)
st.pyplot(fig2)

# Section: Make a prediction
st.header("Make a prediction")
input_data = []
for feature in X.columns:
    val = st.number_input(f"Input {feature}", value=float(X[feature].mean()))
    input_data.append(val)
if st.button("Predict"):
    pred = clf.predict([input_data])[0]
    st.success(f"Prediction: {target_names[pred]}")
```
